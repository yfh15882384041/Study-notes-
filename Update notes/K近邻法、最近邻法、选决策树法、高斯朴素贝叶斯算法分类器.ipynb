```python
#用了#K近邻法、最近邻法、选决策树法、高斯朴素贝叶斯算法分类器
# -*- coding: utf-8 -*-


#构建一个模式识别框架的步骤
#	第一步：数据的加载(getData())
#	第二步：选择模型(somemodel())
#	第三步：模型的训练(model.fit(train_x,train_y))
#	第四步：模型的预测(model.predict(test_x))
#	第五步：模型的评测(score_function(test_y, predictions))
#	第六步：模型的保存(joblib.dump(knn, 'filename.pkl'))

##(1)数据导入，分割数据
#导入digits数据集
from sklearn.datasets import load_digits

#分割数据模块
from sklearn.model_selection import train_test_split

#（2）K最近（KNN，K-Nearest Neighbor）、最近邻法、选决策树法、高斯朴素贝叶斯算法分类器
from sklearn.neighbors import KNeighborsClassifier     #选k近邻法
#from sklearn.neighbors import NearestCentroid         #选最近邻法
#from sklearn.tree import DecisionTreeClassifier       #选决策树法
#from sklearn.naive_bayes import GaussianNB             #高斯朴素贝叶斯




#加载iris数据集
digits = load_digits()
X_digits=digits.data
y_digits=digits.target

 
##分割数据

X_train,X_test,y_train,y_test=train_test_split(X_digits,y_digits,test_size=0.25, random_state=4)

## 建立模型进行训练和预测

#建立模型
knn=KNeighborsClassifier()                 #选k近邻法
#knn=KNeighborsClassifier(n_neighbors=3)  

#knn=NearestCentroid()                      #选最近邻法
#knn=DecisionTreeClassifier()               #选决策树法

#knn=GaussianNB()                             #高斯朴素贝叶斯  



#（3）训练模型
knn.fit(X_train,y_train) #这句有问题
print(knn.score(X_test, y_test)) #计算模型的准确率
print("*"*74)

#（4）预测模型
y_pred=knn.predict(X_test)
print(y_pred)
print("*"*74)
#print(y_pred-y_test) 


###—（5）评价— ——用accuracy_score计算准确率— ———————
import numpy as np
from sklearn.metrics import accuracy_score

print(accuracy_score(y_test, y_pred))#也可以算正确率
print("*"*74)
 
print(accuracy_score(y_test, y_pred, normalize=False)) #统计测试样本正确分类的个数442
print("*"*74)

#（6）保存和加载模型
import joblib
# 用joblib.dump保存模型
joblib.dump(knn, 'KNN77.pkl') 
# # 用joblib.load加载已保存的模型
knn1= joblib.load('KNN77.pkl') 
# #测试读取后的Model
print(knn1.predict(X_digits[0:1]))#预测第一个数据的类别为0，实际是为数字6
print("*"*74)
y_pred1=knn1.predict(X_test)
print(y_pred1) #打印存储的模型再加载进来预测的结果
print("*"*74)
print(y_pred1-y_test) #不为零的数就是被错分的
print("*"*74)

#--------------cross_val_score-----用5折交叉验证-------------------------
from sklearn.model_selection import cross_val_score
scores = cross_val_score(knn, digits.data, digits.target, cv=5)
print(scores)
```
