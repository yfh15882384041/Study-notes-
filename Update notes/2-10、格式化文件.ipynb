## XML(extensibleMarkupLangusge)    可扩展标记语言
  - 标记语言，语言中使用尖括号括起来的文本字符串标记
  - 可扩展，用户可以自己定义需要的标记
  - 例如che
  
          <Teacher>
               自定义标记Teacher
               在两个标记之间任何内容都应该与Teacher有关
          </Teacher>
  - 是w3c组织制定的标准
  - XMl 描述数据本身，即数据的结构和语义
  - Html侧重于如何显示web页面中的数据
 
 
- xml文档的构成
  - 处理指令（可以认为一个文件内只有一个处理指令）
    - 最多只有一行，且必须在第一行
    - 内容是与xml本身处理起先关的一些声明或者指令
    - 以xml开头
    - 一般用于声明xml的版本和采用的编码
      - version属性时必须的
      - encoding属性用来支出xml解释器使用的编码
      - <?xml version="1.0" encoding="utf-8"?>
  - 根元素（一个文件内只有一个根元素）
    - 在整个xml文件中，可以把他看作是一个树形结构
    - 根元素有且只有一个
  - 子元素
  - 属性
  - 内容
    - 表明标签存储的信息
  - 注释
    - 起说明作用的信息
    - 注释不能出现在便签里
    - 只有在注释的开头和结尾使用双短横线
    - 三段横线只能出现在注释的开头而不能出现在注释的结尾
    
- 保留字符的处理
  - xml中使用的符号可能跟实际括号相冲突，典型的就是左右尖括号‘
  - 使用实体引用（EntityReference）来表示保留字符
       
               <scoce> scoce>80 </scoce>  #有错误 xml不能出现
               <scoce> scoce&gt;80</scoce> #允许使用
  - 把含有保留字符的部分放在CDATA块中，CDATA块把内部信息视为不需要转义
  
            <!CDATA[
                select name,age
                from Student
                where score > 80
                ]]>
                
            'i \ love wangxiaojing'
            r'i \ love wangxiaojing'
            
  - 常见的需要转义的保留字符和对应的实体引用
    
            &  &amp;
            <  &lt;
            >  &gt;
            '  &apos;
            "  &quot;
            一共五个都以&开头并且以分号结尾
            
- xml便标的命名规则
  - Pascal命名
  - 用单词的第一个字母大写
  - 大小写严格区分
  - 配对的标签必须一致
    
- 命名空间
  - 为了防止命名冲突
  
       
         <Student>
                <Name>LiuYing</Name>
                <Age>23</Age>
         </Student>
         <Room>
                <Name>2014</Name>
                <Location>1-23-1</Location>
              
         </Room>
         
  - 如果归并上述两个内容消息，会产生冲突
  
         <Schooler>
               <Name>LiuYing</Name>
               <Age>23</Age>
               <Name>2014</Name>
               <Location>1-23-1</Location>
         </Schooler>
  - 为了避免冲突，需要给可能冲突的元素添加命名空间
  - xmlns: xml name space  的缩写
         
         <Schooler> xmlns: student="http:www.my_student" xmlns:room="http:my_room"
               <student:Name>LiuYing</student:Name>
               <Age>23</Age>
               <room:Name>2014</room:Name>
               <Location>1-23-1</Location>
         </Schooler> 
         

### xml访问
- 读取
  - xml读取分为两个主要技术  SAX  DOM
  - SAX（simple API for xml）
    - 基于时间驱动
    - 利用SAX解析文档设计到解析器和事件处理两部分
    - 特点：很快；流式读取
    
  - DOM
    - 是w3c制定的xml编程接口
    - 一个xml文件在缓存中以数结构保存，读取
    - 用途： 1.定位浏览xml任何一个节点信息  2. 添加删除相应的内容
    - minidom： 百度
    - etree：  百度
    
  - xml文件写入
    - ele.set： 修改属性
    - ele.append :添加子元素
    - ele.remove: 删除元素
    - 生成创建
      - subElement：
      - minidom 写入

- JSON格式是一个键值对形式的数据集
  - key：字符串
  - value： 字符串，数字，列表，json

                 "name":"wangdana"
                 "age":18
                 "moblie":"79445671"
                 }
- json和Python格式的对应
  - 字符串：字符串
  - 数字：数字
  - 队列：list
  - 对象：dict
  - 布尔值：布尔值
  
- Python for json

  - json和python对象的转换
